```
# Checking images and their labels
TestNum = 5
diag = 'Melanoma'
diagg = 'NotMelanoma'
PATHtrain='/content/DermMel/train_sep'
for i in range(TestNum):
  image_dir = PATHtrain +'/'+diag+'/'
  img_name = os.listdir(image_dir)[TestNum]
  img_path = image_dir+str(img_name)
  img = image.load_img(img_path, target_size=(224, 224))
  imgplot = plt.imshow(img)
  print( i,"th training data of class ",diag)
  plt.show()
  image_dir = PATHtrain +'/'+diagg+'/'
  img_name = os.listdir(image_dir)[TestNum]
  img_path = image_dir+str(img_name)
  img = image.load_img(img_path, target_size=(224, 224))
  imgplot = plt.imshow(img)
  print( i,"th training data of class ",diagg)
  plt.show()
for i in range(TestNum):
  PATHvalid='/content/DermMel/valid'
  image_dir = PATHvalid +'/'+diag+'/'
  img_name = os.listdir(image_dir)[TestNum]
  img_path = image_dir+str(img_name)
  img = image.load_img(img_path, target_size=(224, 224))
  imgplot = plt.imshow(img)
  print( i,"th validation data of class ",diag)
  plt.show()
  image_dir = PATHvalid +'/'+diagg+'/'
  img_name = os.listdir(image_dir)[TestNum]
  img_path = image_dir+str(img_name)
  img = image.load_img(img_path, target_size=(224, 224))
  imgplot = plt.imshow(img)
  print( i,"th validation data of class ",diagg)
  plt.show()

PATHtest='/content/DermMel/test'
image_dir = PATHtest +'/'+diag+'/'
img_name = os.listdir(image_dir)[TestNum]
img_path = image_dir+str(img_name)
img = image.load_img(img_path, target_size=(224, 224))
imgplot = plt.imshow(img)
print( i,"th test data of class ",diag)
plt.show()
image_dir = PATHtest +'/'+diagg+'/'
img_name = os.listdir(image_dir)[TestNum]
img_path = image_dir+str(img_name)
img = image.load_img(img_path, target_size=(224, 224))
imgplot = plt.imshow(img)
print( i,"th test data of class ",diagg)
plt.show()

#Data Visualization
#plotting some sample images
```
